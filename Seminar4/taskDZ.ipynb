```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['id'], df['price'])
plt.xlabel('Номер наблюдения')
plt.ylabel('Стоимость недвижимости')
plt.title('График стоимости недвижимости')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['sqft_living'], bins=30)
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Количество объектов недвижимости')
plt.title('График распределения квадратуры жилой площади')
plt.show()
```

Изучите распределение квадратуры жилой недвижимости. чем больше квадратура, тем меньше домов.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['yr_built'], bins=30)
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов недвижимости')
plt.title('График распределения года постройки')
plt.show()
```

Изучите распределение года постройки. С каждым годо количество новых обьектов возрастает.

```python
plt.figure(figsize=(6, 4))
df.groupby('waterfront')['id'].count().plot(kind='bar')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.title('График распределения домов от вида на набережную')
plt.xticks([0, 1], ['Нет', 'Да'])
plt.show()
```

очень мало домов у которых  есть вид на набережную

```python
plt.figure(figsize=(10, 6))
plt.hist(df['floors'], bins=30)
plt.xlabel('Этажи')
plt.ylabel('Количество домов')
plt.title('График распределения этажей домов')
plt.show()
```

Изучите распределение этажей домов. много одноэтажных и двухэтажных построек.

```python
plt.figure(figsize=(6, 4))
df.groupby('condition')['id'].count().plot(kind='bar')
plt.xlabel('Состояние домов')
plt.ylabel('Количество домов')
plt.title('График распределения состояния домов')
plt.show()
```


Изучите распределение состояния домов. Больше со средним состоянием домов.
